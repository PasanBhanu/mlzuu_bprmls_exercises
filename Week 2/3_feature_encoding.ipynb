### Importing Libraries

```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Gender -> Nominal

Geography -> Nominal

CreditScoreBins -> Ordinal

```python
df = pd.read_csv('data/processed/ChurnModelling_Binning_Applied.csv')
df.head()
```

### Encode Nominal Features

```python
nominal_variables = ['Geography', 'Gender']

geography_dummies = pd.get_dummies(df['Geography'], prefix='Geography')
gender_dummies = pd.get_dummies(df['Gender'], prefix='Gender')

df_encoded = pd.concat([df, gender_dummies, geography_dummies], axis=1)
del df_encoded['Geography']
del df_encoded['Gender']
df_encoded
```

### Encode Ordinal Features

```python
encode_dict_credit_score = {
    'Poor': 0,
    'Fair': 1,
    'Good': 2,
    'Very Good': 3,
    'Excellent': 4
}

df_encoded['CreditScoreBins'] = df_encoded['CreditScoreBins'].map(encode_dict_credit_score)
df_encoded
```
