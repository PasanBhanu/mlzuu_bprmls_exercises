```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/processed/ChurnModelling_Outliers_Handled.csv')
df = df.drop(df.columns[0], axis=1)
df.head()
```

```python
def custom_binning_credit_score(value):
    if value < 580:
        return 'Poor'
    if value < 670:
        return 'Fair'
    if value < 740:
        return 'Good'
    if value < 800:
        return 'Very Good'
    else:
        assert True, 'Credit score cannot be greater than 850'

df['CreditScoreBins'] = df['CreditScore'].apply(custom_binning_credit_score)
del df['CreditScore']
df.head()
```

```python
df.to_csv('data/processed/ChurnModelling_Binning_Applied.csv', index=False)
```
