```python
import numpy as np
import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE
from matplotlib import pyplot as plt
```

### 1. Load Dataset

```python
df = pd.read_csv('data/raw/data.csv')
df.head()
```

### 2. Build Scikit-learn Pipeline

```python
reminder_features = ['NumOfProducts', 'HasCrCard', 'IsActiveMember', 'Exited']
numerical_features = ['Age', 'Tenure', 'Balance', 'EstimatedSalary']
nominal_features = ['Gender', 'Geography']
ordinal_features = ['CreditScoreBins']
```

```python
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
```

```python
nominal_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('encoder', OneHotEncoder())
])
```

```python
ordinal_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('encoder', OrdinalEncoder())
])
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('nom', nominal_transformer, nominal_features),
        ('ord', ordinal_transformer, ordinal_features)
    ],
    remainder='drop'
)
```

```python
nominal_feature_names = []
for feature in nominal_features:
    unique_values = df[feature].unique()
    nominal_feature_names.extend([f"{feature}_{value}" for value in unique_values])

print(nominal_feature_names)

df_cp = df.copy()
df_transformed = pd.DataFrame(
    preprocessor.fit_transform(df_cp),
    columns=numerical_features + nominal_feature_names + ordinal_features
)

df_transformed
```

```python
df_remainder = df[reminder_features]
df_remainder
```

```python
df_pp = pd.concat([df_transformed, df_remainder], axis=1)
df_pp
```

```python
df_pp.to_csv('data/processed/data_prepared.csv', index=False)
```

### 3. Handle Class Imbalance

```python
X = df_pp.drop(columns=['Exited'])
Y = df_pp['Exited']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
smote = SMOTE(random_state=42)
X_train_resampled, Y_train_resampled = smote.fit_resample(X_train, Y_train)
```

```python
plt.figure(figsize=(15, 4))

plt.subplot(131)
Y_train.value_counts().plot(kind='bar', title='Class Distribution - ')

plt.subplot(132)
Y_train_resampled.value_counts().plot(kind='bar', title='Class Distribution - Resampled')

plt.subplot(133)
Y_test.value_counts().plot(kind='bar', title='Class Distribution - Test')
```

```python
np.savez('artifacts/X_train.npz', X_train=X_train_resampled)
np.savez('artifacts/Y_train.npz', X_train=Y_train_resampled)
np.savez('artifacts/X_test.npz', X_train=X_test)
np.savez('artifacts/Y_test.npz', X_train=Y_test)
```
